```python
import matplotlib.pyplot as plt
```

#### Exercise 01

```python
X = [1, 2, 3, 4]
y = [1, 2, 0, 0.5]

plt.figure(figsize = (10, 5))

plt.scatter(X, y)

plt.title("Gráfico de dispersión")
plt.show()
```

#### Exercise 02

```python
plt.figure(figsize = (10, 5))

plt.plot(X, y)

plt.title("Gráfico de líneas")
plt.show()
```

#### Exercise 03

```python
import numpy as np

np.random.seed(42)

data = np.random.normal(2, 1.5, 1000)

plt.figure(figsize = (10, 5))

plt.hist(data, bins = 30, alpha = 0.7)

plt.title("Histograma")
plt.show()
```

#### Exercise 04

```python
import pandas as pd

dataframe = pd.read_csv("https://raw.githubusercontent.com/cvazquezlos/machine-learning-prework/main/04-matplotlib/assets/titanic_train.csv")
dataframe.head()
```

```python
fig, axis = plt.subplots(2, 1, figsize = (10, 10))

axis[0].hist(dataframe.Age, bins = 30, alpha = 0.7)
axis[1].hist(dataframe.Fare, bins = 15, alpha = 0.7)

plt.tight_layout()
plt.show()
```
